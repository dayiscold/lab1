#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text):
    vowels = set('aeiou')
    count = 0

    for char in text.lower():
        if char in vowels:
            count += 1

    return count


def test_count_vowels():
    assert count_vowels("") == 0
    assert count_vowels("bcdfg") == 0
    assert count_vowels("aeiou") == 5
    assert count_vowels("AEIOU") == 5
    assert count_vowels("hello world") == 3
    assert count_vowels("programming") == 3
    assert count_vowels("Python") == 1
    assert count_vowels("The quick brown fox") == 5

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_count_vowels()

    test_strings = ["hello", "world", "TEST", "aeiou", "xyz", ""]
    for s in test_strings:
        print(f"'{s}': {count_vowels(s)} гласных")
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique_chars(string):
    seen = set()

    for char in string:
        if char in seen:
            return False
        seen.add(char)

    return True


def test_is_unique_chars():
    assert is_unique_chars("") == True
    assert is_unique_chars("a") == True
    assert is_unique_chars("abc") == True
    assert is_unique_chars("hello") == False
    assert is_unique_chars("12345") == True
    assert is_unique_chars("112233") == False
    assert is_unique_chars("Aa") == True

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_is_unique_chars()

    test_cases = ["abcd", "hello", "test", "123", "112", "", "AaBb"]
    for s in test_cases:
        result = is_unique_chars(s)
        print(f"'{s}': {'уникальные символы' if result else 'есть повторения'}")
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    count = 0
    while n > 0:
        if n & 1:
            count += 1
        n >>= 1
    return count


def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(1) == 1
    assert count_bits(2) == 1
    assert count_bits(3) == 2
    assert count_bits(7) == 3
    assert count_bits(8) == 1
    assert count_bits(15) == 4
    assert count_bits(1023) == 10

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_count_bits()

    test_numbers = [0, 1, 2, 3, 7, 8, 15, 255, 1024]
    for num in test_numbers:
        print(f"{num} (бинарное: {bin(num)}) содержит {count_bits(num)} бит(ов) равных 1")
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 10:
        return 0

    count = 0
    while n >= 10:
        product = 1
        while n > 0:
            product *= n % 10
            n //= 10
        n = product
        count += 1

    return count


def test_magic():
    assert magic(4) == 0
    assert magic(39) == 3
    assert magic(999) == 4
    assert magic(10) == 1
    assert magic(25) == 2
    assert magic(123) == 1
    assert magic(1) == 0
    assert magic(777) == 4

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_magic()

    test_numbers = [4, 39, 999, 10, 25, 123, 1, 777]
    for num in test_numbers:
        print(f"{num} -> {magic(num)} шагов")
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")

    total = 0
    n = len(pred)

    for i in range(n):
        diff = pred[i] - true[i]
        total += diff * diff

    return total / n


def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    assert mse([1, 2, 3], [2, 3, 4]) == 1
    assert mse([0, 0, 0], [1, 1, 1]) == 1
    assert mse([1, 2], [3, 4]) == 4
    assert mse([2, 3, 4], [1, 2, 3]) == 1

    try:
        mse([1, 2], [1, 2, 3])
        assert False,
    except ValueError:
        pass

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_mse()

    test_cases = [
        ([1, 2, 3], [1, 2, 3]),
        ([1, 2, 3], [4, 5, 6]),
        ([0, 0], [1, 1]),
        ([2, 4, 6], [3, 5, 7])
    ]

    for pred, true in test_cases:
        result = mse(pred, true)
        print(f"MSE({pred}, {true}) = {result}")
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    if n <= 1:
        return str(n)

    result = ""
    divisor = 2

    while n > 1:
        count = 0
        while n % divisor == 0:
            n //= divisor
            count += 1

        if count > 0:
            if count == 1:
                result += f"({divisor})"
            else:
                result += f"({divisor}**{count})"

        divisor += 1 if divisor == 2 else 2

    return result


def test_prime_factors():
    assert prime_factors(1) == "1"
    assert prime_factors(2) == "(2)"
    assert prime_factors(3) == "(3)"
    assert prime_factors(4) == "(2**2)"
    assert prime_factors(6) == "(2)(3)"
    assert prime_factors(8) == "(2**3)"
    assert prime_factors(12) == "(2**2)(3)"
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(17) == "(17)"
    assert prime_factors(100) == "(2**2)(5**2)"

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_prime_factors()

    test_numbers = [1, 2, 4, 6, 8, 12, 17, 100, 86240]
    for num in test_numbers:
        print(f"{num} = {prime_factors(num)}")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    total = 0
    k = 1

    while total < n:
        total += k * k
        if total == n:
            return k
        k += 1

    return "It is impossible"


def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(91) == 6
    assert pyramid(2) == "It is impossible"
    assert pyramid(3) == "It is impossible"
    assert pyramid(8) == "It is impossible"
    assert pyramid(20) == "It is impossible"

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_pyramid()

    test_numbers = [1, 2, 5, 14, 30, 55, 91, 3, 8, 20]
    for num in test_numbers:
        result = pyramid(num)
        print(f"{num} -> {result}")
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    digits = [int(d) for d in str(n)]
    length = len(digits)

    if length <= 2:
        return True

    mid = length // 2

    if length % 2 == 0:
        left_sum = sum(digits[:mid - 1])
        right_sum = sum(digits[mid + 1:])
    else:
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid + 1:])

    return left_sum == right_sum


def test_is_balanced():
    assert is_balanced(1) == True
    assert is_balanced(11) == True
    assert is_balanced(121) == True
    assert is_balanced(1234) == False
    assert is_balanced(123321) == True
    assert is_balanced(12345) == False
    assert is_balanced(1234321) == True
    assert is_balanced(1221) == True
    assert is_balanced(12321) == True
    assert is_balanced(123454321) == True
    assert is_balanced(123) == False

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_is_balanced()

    test_numbers = [1, 11, 121, 1234, 123321, 12345, 1221, 123, 1234321]
    for num in test_numbers:
        result = is_balanced(num)
        print(f"{num} -> {'сбалансировано' if result else 'несбалансировано'}")
```
